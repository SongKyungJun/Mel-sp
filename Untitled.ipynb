```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import IPython.display
import librosa.display
```

```python
Song3 = np.load("./arena_mel_0/arena_mel/0/7.npy")
Song3.shape
```

```python
plt.figure(figsize=(20, 10))
librosa.display.specshow(librosa.power_to_db(Song3, ref=np.max), y_axis='mel', sr=16000, hop_length=256, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-Spectrogram')
plt.tight_layout()
plt.savefig('Mel-Spectrogram example.png')
plt.show()
#sr=16000인 이유는 애초에 생성할때 30초짜리를 16000
```

```python
mfcc = librosa.feature.mfcc(S=Song3, sr=16000, n_fft=512, hop_length=256, n_mfcc=13)
```

```python
delta2_mfcc = librosa.feature.delta(mfcc, order=2)
```
