```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome('C:/Users/Song Kyung Jun/Desktop/skj/jpnt/Min_PJ2/chromedriver')
driver.get("https://www.melon.com/dj/chart/djchart_list.htm")
```

```python
k = list(range(0,21))
k2=[]
for i in k:
    k1 = str(k[i])
    k2.append(k1)
del k2[0]
```

```python
path0 = '//*[@id="djPlylstList"]/div/ul/li['
path1 = k2
path2 = ']/div[2]/div[2]/button'
path3 = []
for i in k2:
    p = path0 + i + path2
    path3.append(p)
path3[0]
```

```python
plts=[]
for z in path3:
    pltitle = driver.find_element_by_xpath(z)
    plts.append(pltitle)
    
pltitle2=[]
for i in plts:
    pltitle2.append(i.get_attribute('data-djcol-no'))
```

```python
pl_list_data = pd.DataFrame({'Playlist':[],
                             'PlTag':[],
                             'Plmadename':[],
                             'PlLcnt':[],
                             'Song':[],
                             'Artist':[],
                             'Album':[],
                            'Sncnt':[]})


for i in pltitle2:
    driver.get("https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq="+i)
    title=driver.find_elements_by_class_name('ellipsis.rank01')
    stitle2=[]
    for j in title:
        stitle2.append(j.text)
    
    singer=driver.find_elements_by_class_name('ellipsis.rank02')
    singer2=[]
    for j in singer:
        singer2.append(j.text)
    
    album=driver.find_elements_by_class_name('ellipsis.rank03')
    album2 = []
    for j in album:
        album2.append(j.text)
    
    plname = driver.find_element_by_class_name('ellipsis.song_name')
    plname2 = []
    plname2.append(plname.text)
    plname3 = plname2*len(album2)
    
    plct = driver.find_element_by_xpath('//*[@id="conts"]/div[2]/div/div[2]/div[4]/button[1]/span[2]')
    plct2 = []
    plct2.append(plct.text)
    plct3 = plct2*len(album2)
    
    tag=driver.find_element_by_class_name('tag_list.type03')
    tag2=[]
    tag2.append(tag.text)
    tag3 = tag2*len(album2)
    
    date=driver.find_element_by_class_name('date')
    date2=[]
    date2.append(date.text)
    date3 = date2*len(album2)
    
    
    t = len(album2)
    t2=[]
    for i in range (1,t+1):
        t1 = i
        t2.append(t1)
    t3 = list(map(str,t2))

    pt3 = []
    pt0=('//*[@id="frm"]/div/table/tbody/tr[')
    pt1=t3
    pt2 = (']/td[7]/div/button/span[2]')
    for v in t3:
        p = pt0 + v + pt2
        pt3.append(p)
        
    Slct2 = []    
    for y in pt3:
        Slct=driver.find_elements_by_xpath(y)
        for j in Slct:
            Slct2.append(j.text)
    
    
    add_data = {'Playlist':plname3,'PlTag':tag3,'Plmadename':date3,
                'PlLcnt':plct3,'Song':stitle2, 'Artist':singer2,
                'Album':album2, 'Sncnt':Slct2}
    add_data_df = pd.DataFrame(add_data)
    
    pl_list_data = pl_list_data.append(add_data_df, ignore_index = True)
    
```
