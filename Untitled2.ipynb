```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome('C:/Users/Administrator/jpnt/Min_Pj/chromedriver')
driver.get("https://www.melon.com/dj/essential/djessential_list.htm?depth1TagSeq=133&depth2TagSeq=ALL")
```

```python
k = list(range(0,21))
k2=[]
for i in k:
    k1 = str(k[i])
    k2.append(k1)
del k2[0]
```

```python
path0 = '//*[@id="djPlylstList"]/div/ul/li['
path1 = k2
path2 = ']/div[2]/div[2]/button'
path3 = []
for i in k2:
    p = path0 + i + path2
    path3.append(p)
path3[0]
```

```python
plts=[]
for z in path3:
    pltitle = driver.find_element_by_xpath(z)
    plts.append(pltitle)
    
pltitle2=[]
for i in plts:
    pltitle2.append(i.get_attribute('data-djcol-no'))
```

```python
pl_list_data = pd.DataFrame({'Playlist':[],
                             'PlTag':[]})


for i in pltitle2:
    driver.get("https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq="+i)
    
    plname = driver.find_element_by_class_name('ellipsis.song_name')
    plname2 = []
    plname2.append(plname.text)

    tag=driver.find_element_by_class_name('tag_list.type03')
    tag2=[]
    tag2.append(tag.text)

    
    
    add_data = {'Playlist':plname2,'PlTag':tag2}
    add_data_df = pd.DataFrame(add_data)
    
    pl_list_data = pl_list_data.append(add_data_df, ignore_index = True)
    
```
